7. Create a vector with values ranging from 10 to 49

```python
v = np.arange(10, 50)
print(v)
```

8. Reverse a vector (first element becomes last)

```python
print("Previous : \n", v)

v[:] = v[::-1]
print("Reversed with v[:] = v[::-1] :\n", v)

u = np.flip(v)
print("With np.flip(v)\n", u)
```

10. Find indices of non-zero elements from [1,2,0,0,4,0]

```python
v = [1,2,0,0,4,0]
v = np.array(v)
r = []
for i in range(v.size):
    r.append(i)
    print(f"{i}, ") if v[i] == 0 else None
```
