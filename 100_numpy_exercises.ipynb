18. Create a 5x5 matrix with values 1,2,3,4 just below the diagonal 

````python
v = np.random.random((5,5))*10
c = np.array([1, 2, 3, 4])
for i in range(c.size):
    for j in range(c.size):
        if i == j:
            v[i+1, j] = c[j] 

print(v)
````

19. Create a 8x8 matrix and fill it with a checkerboard pattern 

````python
v = np.ones(64)
for i in range(64):
    if i%2 == 0:
        v[i] = 0
v = v.reshape(8,8)
print(v)
````

20. Consider a (6,7,8) shape array, what is the index (x,y,z) of the 100th element?

````python
# Total elements
shape = (6, 7, 8)
total = np.prod(shape)
print(f"Total elements: {total}") 

# Flatten and find the 100th element (index 99 in 0-based)
flat_index = 99  # 0-based index for 100th element

# Method 1: Using np.unravel_index()
index = np.unravel_index(flat_index, shape)
print(f"Index (x,y,z): {index}") 
````
